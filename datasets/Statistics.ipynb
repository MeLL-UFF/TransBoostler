```python
experiments = [
            {'dataset':'imdb','predicate':'workedunder'},
            {'dataset':'uwcse','predicate':'advisedby'},
            {'dataset':'cora','predicate':'samevenue'},
            {'dataset':'yeast','predicate':'proteinclass', 'recursion': True},
            {'dataset':'twitter','predicate':'accounttype', 'recursion': True},
            {'dataset':'nell_sports','predicate':'teamplayssport'},
            {'dataset':'nell_finances','predicate':'companyeconomicsector'}
]
```

## Process

```python
table = []


for exp in experiments:
    dataset = exp['dataset']
    target_predicate = exp['predicate']
    
    dt = datasets.load(dataset, bk[dataset], target_predicate)
    number_predicates = len(set([pred.split('(')[0] for pred in bk[dataset]]))
    
    if 'recursion' in exp:
        number_predicates -= 1
    
    constants = []
    unique_constants = []
    unique_types = []
    true_ground_constants = []
    
    facts_ground_literals = datasets.group_folds(dt[0])
    true_ground_literals = datasets.group_folds(dt[1])
    false_ground_literals = datasets.group_folds(dt[2])
    
    types = [re.search(r'\((.*?)\)', i).group(1).split(',') for i in bk[dataset]]
    
    for _type in types:
        unique_types += [t.replace('-','').replace('+', '').replace('`', '') for t in _type]
    
    number_true_ground_literals = len(true_ground_literals)
    number_false_ground_literals = len(false_ground_literals)
    true_ground_constants += [re.search(r'\((.*?)\)', i).group(1).split(',') for i in true_ground_literals]
    
    constants += [re.search(r'\((.*?)\)', i).group(1).split(',') for i in facts_ground_literals]
    constants += [re.search(r'\((.*?)\)', i).group(1).split(',') for i in false_ground_literals]
    constants += true_ground_constants
    
    for c in constants:
        unique_constants += c
    
    type_1, type_2 = [], []
    for c in true_ground_constants:
        type_1 += c[0]
        type_2 += c[1]

    table.append([dataset,
                 len(set(unique_constants)), 
                 len(set(unique_types)),
                 number_predicates,
                 number_true_ground_literals,
                 number_false_ground_literals,
                 len(set(type_1))*len(set(type_2))*number_true_ground_literals
                 ])
    
display(pd.DataFrame(table, columns=['Dataset', 'Number of Constants', 'Number of Types', 'Number of Predicates', 'Number of positives examples', 'Number of negative examples', 'Total number of ground literals']))
```
