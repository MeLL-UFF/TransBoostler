## Leitura e Processamento dos Dados

```python
def read_transfer_file(filename):
        mappings = {}
        with open(os.getcwd() + '/mapeamentos/' + filename, 'r') as file:
            for line in file:
                if('param' in line.lower() or 'map' in line.lower()): continue
                line = line.replace('\n', '').split(':')
                source, target = line[0].split('(')[0].strip(), line[1].split('(')[0].strip()
                mappings[source] = target
        return mappings
```

```python
filename = '{}/{}/transfer_fasttext_{}.txt'
for experiment in experiments: 
    experiment_title = str.format('{}_{}_{}', experiment['id'], experiment['source'], experiment['target'])
    
    display(Markdown('# Results for ' + experiment_title))
    
    for metric in ['euclidean', 'softcosine', 'wmd', 'relax-wmd']:
        table = []
        
        display(Markdown('## ' + metric))
        
        stopwords_normalized = read_transfer_file(filename.format('mapeamentos-com-stopwords-normalizados', experiment_title, metric))
        stopwords_no_normalized = read_transfer_file(filename.format('mapeamentos-com-stopwords-sem-normalizacao', experiment_title, metric))
        no_stopwords_normalized = read_transfer_file(filename.format('mapeamentos-sem-stopwords-normalizados', experiment_title, metric))
        no_stopwords_no_normalized = read_transfer_file(filename.format('mapeamentos-sem-stopwords-sem-normalizacao', experiment_title, metric))
        
        for source in stopwords_normalized.keys():
            table.append([source,stopwords_no_normalized[source],no_stopwords_no_normalized[source],stopwords_normalized[source],no_stopwords_no_normalized[source]])
        
        display(pd.DataFrame(table, columns=['source', '+stopwords', '-stopwords', '+stopwords+normalization','-stopwords-normalization']))
```
